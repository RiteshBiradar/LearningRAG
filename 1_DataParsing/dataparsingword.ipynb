```python
from langchain_community.document_loaders import Docx2txtLoader, UnstructuredWordDocumentLoader
```

```python
wordLoader = Docx2txtLoader("data/word/sample.docx")
wordDocuments = wordLoader.load()
print(f"{wordDocuments}")
```

```python
unstructuredLoader = UnstructuredWordDocumentLoader("data/word/sample.docx",mode="elements")
unstructuredDocuments = unstructuredLoader.load()
print(f"{unstructuredDocuments}")
```
